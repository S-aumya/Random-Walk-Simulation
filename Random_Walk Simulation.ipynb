```python
import numpy as np
import matplotlib.pyplot as plt

# Set the number of steps
n_steps = 1000

# Generate random steps (either +1 or -1)
steps = np.random.choice([-1, 1], size=n_steps)

# Calculate the cumulative sum of steps
cumulative_sum = np.cumsum(steps)

# Create an array of time points
time_points = np.arange(n_steps)

# Plot the random walk
plt.figure(figsize=(10, 6))
plt.plot(time_points, cumulative_sum, label="Random Walk")
plt.xlabel("Time Steps")
plt.ylabel("Position")
plt.title("Random Walk Simulation")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

def random_walk_2d(n):
    # Initialize starting position
    x, y = 0, 0

    # Lists to store coordinates
    x_coords = [x]
    y_coords = [y]

    # Define possible steps (up, down, left, right)
    steps = [(0, 1), (0, -1), (1, 0), (-1, 0)]

    # Perform n random jumps
    for _ in range(n):
        dx, dy = random.choice(steps)
        x += dx
        y += dy
        x_coords.append(x)
        y_coords.append(y)

    return x_coords, y_coords

# Set lattice parameter (a)
a = 1

# Number of jumps
n_jumps = 1000

# Generate random walk coordinates
x_coords, y_coords = random_walk_2d(n_jumps)

# Scale coordinates by lattice parameter
x_scaled = [a * x for x in x_coords]
y_scaled = [a * y for y in y_coords]

# Plot the random walk
plt.figure(figsize=(6, 6))
plt.plot(x_scaled, y_scaled, marker='o', markersize=2, linestyle='-', color='b')
plt.title("Random Walk on a Square Lattice (2D)")
plt.xlabel("x (scaled)")
plt.ylabel("y (scaled)")
plt.grid(True)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Initialize starting position
x, y = 0, 0

# Lists to store coordinates
x_coords = [x]
y_coords = [y]

# Set lattice parameter (a)
a = 1

# Number of steps (each step consists of 10 jumps)
n_steps = 100

# Perform n_steps random steps
for _ in range(n_steps):
    for _ in range(10):  # 10 jumps per step
        # Randomly decide whether to change direction
        prob_change = random.random()

        # Change direction in x
        if prob_change < 0.25:
            x += 1
        elif 0.25 <= prob_change < 0.5:
            x -= 1

        # Change direction in y
        elif 0.5 <= prob_change < 0.75:
            y += 1
        else:
            y -= 1

        x_coords.append(x)
        y_coords.append(y)

# Scale coordinates by lattice parameter
x_scaled = [a * x for x in x_coords]
y_scaled = [a * y for y in y_coords]

# Plot the random walk
plt.figure(figsize=(6, 6))
plt.plot(x_scaled, y_scaled, marker='o', markersize=2, linestyle='-', color='b')
plt.title("Random Walk on a Square Lattice (2D) with Custom Probabilities")
plt.xlabel("x (scaled)")
plt.ylabel("y (scaled)")
plt.grid(True)
plt.show()
```
